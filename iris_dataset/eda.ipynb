# EDA - Iris Flower Dataset

```
The Iris dataset was used in R.A. Fisher's classic 1936 paper, The Use of Multiple Measurements in Taxonomic Problems, and can also be found on the UCI Machine Learning Repository.

It includes three iris species with 50 samples each as well as some properties about each flower. One flower species is linearly separable from the other two, but the other two are not linearly separable from each other.

The columns in this dataset are:

Id
SepalLengthCm
SepalWidthCm
PetalLengthCm
PetalWidthCm
Species
```

### Numerical Columns

````python
output = plt.boxplot(df_original["SepalLengthCm"], meanline=True, showmeans=True)
print(output.keys())
info = df_original["SepalLengthCm"].describe()
median = df_original["SepalLengthCm"].median()
mean = df_original["SepalLengthCm"].mean()
for line in output["medians"] :
    x, y = line.get_xydata()[1]
    plt.annotate(f"Median={median}", [x+0.01, y])

for line in output["means"] :
    x, y = line.get_xydata()[0]
    plt.annotate(f"Mean={mean:.2f}", [x-0.17, y])

plt.grid(True)
````

````python
def boxplots(dataframe: pd.DataFrame, numeric_features: list, target: str, ncols: int=5, nrows: int=5, figsize: tuple=(10,10)) :
    """
     Generates a box plot for all the numeric features present in your dataframe
    """

    fig = plt.figure(figsize=figsize)

    for i, col in enumerate(numeric_features) :
        ax = fig.add_subplot(nrows, ncols, i+1)
        sns.boxplot(data=dataframe, x=col, hue=target, orient='h', ax=ax)

    fig.suptitle("Box Plots of Numeric Columns")
    fig.tight_layout()
    return fig
````

````python
fig_box = boxplots(df_original, numeric_col, target=target_col, nrows=2, ncols=2, figsize=(20,15))
fig_box.savefig("BoxPlots.jpg")
````

````python
def histplots(dataframe: pd.DataFrame, numeric_features: list, target: str, ncols: int=5, nrows: int=5, figsize: tuple=(10,10)) :
    """
     Generates a box plot for all the numeric features present in your dataframe
    """

    fig = plt.figure(figsize=figsize)

    for i, col in enumerate(numeric_features) :
        ax = fig.add_subplot(nrows, ncols, i+1)
        sns.histplot(data=dataframe, x=col, ax=ax, kde=True, color="blue", hue=target)

    fig.suptitle("Distribution of Numeric Columns")
    fig.tight_layout()
    return fig
````

````python
fig_hist = histplots(df_original, numeric_col, nrows=2, ncols=2, target=target_col, figsize=(20,15))
fig_hist.savefig("HistPlots.jpg")
````

````python
def violinplots(dataframe: pd.DataFrame, numeric_features: list, target: str, ncols: int=5, nrows: int=5, figsize: tuple=(10,10)) :
    """
     Generates a box plot for all the numeric features present in your dataframe
    """

    fig = plt.figure(figsize=figsize)

    for i, col in enumerate(numeric_features) :
        ax = fig.add_subplot(nrows, ncols, i+1)
        sns.violinplot(data=dataframe, x=col, hue=target, orient='h', ax=ax)
        ax.set_title(col, fontsize=20)
        ax.set_xlabel("")

    fig.suptitle("Box Plots of Numeric Columns", fontsize=24)
    fig.tight_layout()
    return fig
````

````python
fig_violin = violinplots(df_original, numeric_col, target_col, 2, 2, figsize=(20,15))
fig_violin.savefig("ViolinPlots.jpg")
````

````python
def stripplots(dataframe: pd.DataFrame, numeric_features: list, target: str, ncols: int=5, nrows: int=5, figsize: tuple=(10,10)) :
    """
     Generates a box plot for all the numeric features present in your dataframe
    """

    fig = plt.figure(figsize=figsize)

    for i, col in enumerate(numeric_features) :
        ax = fig.add_subplot(nrows, ncols, i+1)
        sns.stripplot(data=dataframe, x=col, hue=target, orient='h', ax=ax)
        ax.set_title(col, fontsize=20)
        ax.set_xlabel("")

    fig.suptitle("Box Plots of Numeric Columns", fontsize=24)
    fig.tight_layout()
    return fig
````

````python
fig_strip = stripplots(df_original, numeric_col, target_col, 2, 2, figsize=(20,15))
fig_strip.savefig("StripPlots.jpg")
````

## Outlier Detection

### Univariate Detection

````python
boxplots(df_original, numeric_col, target_col, figsize=(20,20))
````

````python
from scipy.stats import zscore
````

````python
df_petal_len = pd.DataFrame(zscore(df_original["PetalLengthCm"]))
df_petal_len.columns = ["petal_len"]
````

````python
df_original[df_petal_len["petal_len"].abs() > 3].shape
````

````python
df_outlier = pd.DataFrame()
````

````python
for col in numeric_col :
    df_outlier[col] = pd.DataFrame(zscore(df_original[col]))
    r, c = df_original[df_outlier[col].abs() > 3].shape
    print(f"Outliers Count ({col}): {r}")
````

### Multivariate Detection

````python
from sklearn.ensemble import IsolationForest

df_outlier["Outlier"] = IsolationForest(random_state=42).fit_predict(df_original[numeric_col])
df_outlier[df_outlier["Outlier"] == -1].shape
````

````python
df_area = pd.DataFrame()
````

````python
df_area["Petal_Area"] = df_original["PetalLengthCm"] * df_original["PetalWidthCm"]
df_area["Sepal_Area"] = df_original["SepalLengthCm"] * df_original["SepalWidthCm"]
df_area["Target"] = df_original[target_col]
````
